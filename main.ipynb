```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
res = 10


x = np.linspace(-res, res, res*2+1)
y = np.linspace(-res, res, res*2+1)
(x, y) = np.meshgrid(x, y)
# Make variable u a sine function bounded by the res variable
u = np.sin(x / res * np.pi)
v = np.cos(y / res * np.pi)

fig, ax = plt.subplots(figsize = (7, 7))
ax.set_title('Wind Vector Field')
ax.set_xticks(np.arange(-res, res + 1, 1))
ax.set_yticks(np.arange(-res, res + 1, 1))
ax.grid(True)

## Plot random dot
x_pos = np.random.randint(-res, res + 1)
y_pos = np.random.randint(-res, res + 1)
u_val = np.sin((x_pos / res * np.pi))
v_val = np.cos((y_pos / res * np.pi))
speed = np.sqrt(u_val**2 + v_val**2)
direction = np.arctan2(v_val, u_val) * 180 / np.pi

## Round speed and direction to 2 decimal places.
speed = round(speed, 2)
direction = round(direction, 2)


print("Position: (" + str(x_pos) + ", " + str(y_pos) + ") Direction: " + str(direction) +
      " Speed: " + str(speed))
ax.plot(x_pos, y_pos, 'ro', markersize = 8)


# Size of x, y and u, v must be the same
plt.quiver(x, y, u, v, scale_units= 'xy', scale= 1)
plt.show()
```
