```python
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
import numpy as np
```

```python
# Create a grid for chosen resolution.
res = 10
x = np.linspace(-res, res, (res * 2) + 1)
y = np.linspace(-res, res, (res * 2) + 1)
X, Y = np.meshgrid(x, y)
#print(X)

U = 1.0 + 0.3 * np.sin(Y * 2) + 0.1 * np.random.randn((res * 2) + 1, (res * 2) + 1)
V = 0.2 * np.sin(X * 3) + 0.1 * np.random.randn((res * 2) + 1, (res * 2) + 1)

# Normalize the vectors.
magnitude = np.sqrt(U**2 + V**2)
U /= magnitude.max()
V /= magnitude.max()
```

```python
# Create figure and its trai
fig, ax = plt.subplots(figsize = (res/2, res/2))
ax.set_title('Wind Vector Field')
ax.set_xticks(np.arange(-res, res + 1, 1))
ax.set_yticks(np.arange(-res, res + 1, 1))
ax.grid(True, linestyle = '--', color = 'gray', linewidth = 0.5)
ax.set_xticks(np.arange(-res, res + 1, 2), minor=True)
ax.set_yticks(np.arange(-res, res + 1, 2), minor=True)

# Plot random dot.
x_pos = np.random.randint(0, 6)
y_pos = np.random.randint(0, 6)

# Look over again! Makes no sense
# u_val = np.sin((x_pos / res * np.pi))
# v_val = np.cos((y_pos / res * np.pi)) 

# speed = np.sqrt(u_val**2 + v_val**2)

# Double check calculation
# direction = np.arctan2(v_val, u_val) * 180 / np.pi

# Round speed and direction to 2 decimal places.
# speed = round(speed, 2)
# direction = round(direction, 2)

# Print related data.
# print("Position: (" + str(x_pos) + ", " + str(y_pos) + ") Direction: " + str(direction) +
#       " Speed: " + str(speed))
ax.plot(x_pos, y_pos, 'ro', markersize = 8)


# Size of x, y and u, v must be the same
plt.quiver(X, Y, U, V, scale_units='xy')
plt.show()
```
