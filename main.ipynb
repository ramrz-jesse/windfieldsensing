## Imports ##

```python
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
import numpy as np
```

```python
# Create a grid for chosen resolution
res = 32
x = np.linspace(0, 2 * np.pi, res)
y = np.linspace(0, 2 * np.pi, res)
X, Y = np.meshgrid(x, y)

U = 1.0 + 0.3 * np.sin(Y * 2) + 0.1 * np.random.randn(res, res)

V = 0.2 * np.sin(X * 3) + 0.1 * np.random.randn(res, res)

# Normalize the vectors for better visualization
magnitude = np.sqrt(U**2 + V**2)
U /= magnitude.max()
V /= magnitude.max()
```

```python
fig, ax = plt.subplots(figsize = (7, 7))
ax.set_title('Wind Vector Field')
ax.set_xticks(np.arange(-res, res + 1, 1))
ax.set_yticks(np.arange(-res, res + 1, 1))
ax.grid(True)

# Plot random dot to check if the wind vector field is correct
x_pos = np.random.randint(0, 6)
y_pos = np.random.randint(0, 6)
# u_val = np.sin((x_pos / res * np.pi))
# v_val = np.cos((y_pos / res * np.pi))
# speed = np.sqrt(u_val**2 + v_val**2)
# direction = np.arctan2(v_val, u_val) * 180 / np.pi

# # Round speed and direction to 2 decimal places.
# speed = round(speed, 2)
# direction = round(direction, 2)


# print("Position: (" + str(x_pos) + ", " + str(y_pos) + ") Direction: " + str(direction) +
#       " Speed: " + str(speed))
ax.plot(x_pos, y_pos, 'ro', markersize = 8)


# Size of x, y and u, v must be the same
plt.quiver(X, Y, U, V, scale_units='xy')
plt.show()
```
